```python
import numpy as np
np.random.seed(1018)
import matplotlib.pyplot as plt
import random as rm
import pandas as pd
from sklearn import preprocessing
from keras.layers import TimeDistributed,Dense,GRU,Dropout,LSTM,Activation
from keras.utils import to_categorical
from keras.models import Sequential
from keras.optimizers import Adam,RMSprop


TIME_STEPS = 10
BATCH_SIZE = 1
INPUT_SIZE = 1
OUTPUT_SIZE = 1
CELL_SIZE = 128
```

```python
x = np.arange(0,400,0.1)
indice = x.shape[0]
y = rm.uniform(-0.5,0.5)+np.sin(x)
dataset = np.round(y,3).reshape(-1,1)

plt.figure(figsize=(100,30))
plt.xlim(x.min()*1.1,x.max()*1.1,)
plt.ylim(y.min()*1.1,y.max()*1.1)
plt.plot(x,y,"bo-",linewidth=2, markersize=5)
plt.show()
trainset = dataset[0:3000]
testset = dataset[3000:4000]
```

```python
def GetBatch(data,BATCH_SIZE,TIME_STEPS):
    data_X = []
    data_y = []
#    for INDEX in range(0,len(data)-2*TIME_STEPS*BATCH_SIZE,TIME_STEPS*BATCH_SIZE):
    for INDEX in range(0,len(data)-TIME_STEPS*BATCH_SIZE):
        X_step = data[INDEX:INDEX+TIME_STEPS*BATCH_SIZE]
#         y_step = data[INDEX+TIME_STEPS*BATCH_SIZE:INDEX+TIME_STEPS*BATCH_SIZE*2]
        y_step = data [INDEX+TIME_STEPS*BATCH_SIZE]
        data_X.append(X_step)
        data_y.append(y_step)
    return np.array(data_X),np.array(data_y)
```

```python
trainX,trainy = GetBatch(trainset,BATCH_SIZE,TIME_STEPS)
testX,testy = GetBatch(testset,BATCH_SIZE,TIME_STEPS)
```

```python
model = Sequential()
model.add(LSTM(CELL_SIZE ,input_shape =(TIME_STEPS,INPUT_SIZE)))
# model.add(LSTM(units=CELL_SIZE))
# model.add(LSTM(units=CELL_SIZE,return_sequences=True))
model.add(Dense(10,activation='relu'))
model.add(Dense(OUTPUT_SIZE))
optimizer = Adam()
model.compile(optimizer=optimizer,loss='mse')
```

```python
model.fit(trainX,trainy,batch_size=BATCH_SIZE,epochs= 2)
```

```python
def prediction(model, batch_size,time_step,X_test):
#     X_test = X_test[0:-1]
#     X_test = np.array(X_test).reshape(batch_size, time_step,1)
    y_pred = model.predict(X_test, batch_size=batch_size)
    return y_pred
```
